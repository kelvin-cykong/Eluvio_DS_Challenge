### Title Text Pre-Processing

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
```

```python
lemma=WordNetLemmatizer() 
stopwords_en = stopwords.words('english')

def text_preprocess (sentence):
    text = sentence['title']
    text = re.sub(r"\'r", "", text)
    text = re.sub(r"\'n", "", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r",", " ", text)
    text = re.sub(r"\.", "", text)
    text = re.sub(r"\^", " ^ ", text)
    text = re.sub(r"U of", "University of", text)
    text = re.sub(r"N Korea", "NorthKorea", text)
    text = re.sub(r"'", " ", text)
    text = re.sub(r"(\d+)(k)", r"\g<1>000", text)
    text = re.sub(r"@\S+", "", text)

    word=nltk.word_tokenize(text.lower())
    new_word=[w for w in word if w not in stopwords_en and w.isalpha()]
    new_word=[lemma.lemmatize(w, "v") for w in new_word]
    sentence = (" ").join(new_word)
    return sentence
```

## Algorithms for Classification

### Preparing Dataset for Training and Testing

```python
from random import sample
import random
random.seed(2021)
data_filtered = data[data['up_votes']<=np.percentile(data['up_votes'], 100)]
data_sampled = data_filtered 
len(data_sampled)
```

```python
data_sampled['attractive_news'].value_counts()
```

```python
title = data_sampled.apply(lambda sentence: text_preprocess(sentence), axis=1).values
label = data_sampled['attractive_news'].values
```

```python
# Without K-fold Cross-Validation
from sklearn.model_selection import train_test_split
title_train, title_test, label_train, label_test = train_test_split(title, label, test_size=0.25, random_state=12312541)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(1,3))
vectorizer.fit(title_train)

title_train = vectorizer.fit_transform(title_train)
title_test = vectorizer.transform(title_test)
title_train
```

### Testing using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(solver='lbfgs', max_iter=10000000)
classifier.fit(title_train, label_train)
score = classifier.score(title_test, label_test)
```

```python
print("Accuracy", score)
```

### Building Simple Neural Network

```python
#Prevent Tensorflow Blas GEMM Launch Failed
import tensorflow as tf
configuration = tf.compat.v1.ConfigProto()
configuration.gpu_options.allow_growth = True
session = tf.compat.v1.Session(config=configuration)
```

```python
from keras.models import Sequential
from keras import layers
from keras.layers import Dense
from keras.layers import Dropout

input_dim = title_train.shape[1]

model = Sequential()
model.add(layers.Dense(16, input_dim=input_dim, activation='relu'))
model.add(Dropout(0.1))
model.add(layers.Dense(64, activation='relu'))
model.add(Dropout(0.1))
model.add(layers.Dense(32, activation='relu'))
model.add(Dropout(0.1))
model.add(layers.Dense(8, activation='relu'))
model.add(Dropout(0.1))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])
model.summary()
```

```python
# Visualizing The Neural Network
import visualkeras

visualkeras.layered_view(model)
```

```python
# Clear session for rerun the model.
from keras.backend import clear_session
clear_session()
```

```python
history = model.fit(title_train, label_train,
                    epochs=25,
                    verbose=False,
                    validation_data=(title_test, label_test),
                    batch_size=256)
```

```python
loss, accuracy = model.evaluate(title_train, label_train, verbose=False)
print("Training Accuracy: {:.4f}".format(accuracy))
loss, accuracy = model.evaluate(title_test, label_test, verbose=False)
print("Testing Accuracy:  {:.4f}".format(accuracy))
```

#### Visualizing Model Accuracy and Loss

```python
# Plotting training and testing accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Training and Validation Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Training and Validation Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
